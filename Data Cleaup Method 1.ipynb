```python
#import libraries
import pandas as pd
import numpy as np
import re
import nltk

#import Stopwords from nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

pd.set_option('display.max_colwidth',240)
```

```python
#Import and Read Data file
DATA_FILE = 'user_reviews.csv'
reviews = pd.read_csv(DATA_FILE, usecols=['text'])
reviews.columns = ['text']
```

```python
#View Text Column
reviews.head(15)
```

```python
Space_replace = re.compile('[/(){}\[\]\|@,;]')
Remove_Symbols = re.compile('[^0-9a-z #+_]')
Stopwords = set(stopwords.words('english'))

def text_prepare(text):
    text = text.lower()
    text = Space_replace.sub(" ",text)
    nosymb = Remove_Symbols.sub("",text)
    text = ' '.join([word for word in nosymb.split() if word not in Stopwords])
    return text
```

```python
#Apply text_prepare function on Review column
reviews.text = reviews.text.apply(text_prepare)
```

```python
#Save the Cleaned Text
reviews.to_csv('Cleaned_Reviews_data.csv',encoding='utf-8')
```
