```python
#import libraries
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_colwidth',240)
```

```python
#Import and Read Data file
DATA_FILE = 'user_reviews.csv'
reviews = pd.read_csv(DATA_FILE, usecols=['text'])
reviews.columns = ['text']
```

```python
#View Text Column
reviews.head(15)
```

```python
#turn text to lower case and remove all special characters
def remove_handles(review):
    return re.sub(r'([^\s\w]|_)+', ' ', review.lower())
```

```python
#Apply remove_handles function on Review column
reviews.text = reviews.text.apply(remove_handles)
```

```python
reviews.head(15)
```

```python
#Extend all Contractions 
def clean(review):
    return review.replace('hadn t' , 'had not')\
        .replace('wasn t', 'was not')\
        .replace('didn t', 'did not')\
        .replace('is t', 'is not')\
        .replace('it s', 'it is')\
        .replace('weren t', 'were not')\
        .replace('i m', 'i am')\
        .replace('don t', 'do not')\
        .replace('can t', 'cannot')\
        .replace('haven t', 'have not')\
        .replace('wouldn t', 'would not')\
        .replace('doesn t', 'does not')\
        .replace('you re', 'you are')\
        .replace('couldn t', 'could not')\
        .replace('let s', 'let us')\
        .replace('there s', 'there is')\
        .replace('we re', 'we are')\
        .replace('i ve', 'i have')\
        .replace('we ve', 'we have')\
        .replace('they ve', 'they have')\
        .replace('i d', 'i would')\
        .replace('you d', 'you would')\
        .replace('you ll', 'you will')\
        .replace('won t', 'will not')\
        .replace('it ll', 'it will')\
        .replace('aren t', 'are not')\
        .replace('here s', 'here is')\
        .replace('shouldn t', 'should not')\
        .replace('there ll', 'there will')\
        .replace('they ll', 'they will')
```

```python
#Apply clean function on Review column
reviews.text = reviews.text.apply(clean)
```

```python
#Save the Cleaned Text
reviews.to_csv('Cleaned_Reviews_data.csv',encoding='utf-8')
```
